**# Check Missing Values**

*   Identifies columns with missing data
*   Prevents bias and errors during analysis



```python
import numpy as np

# Identify column types
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include="object").columns

# Missing values BEFORE
print("Missing values before handling:\n")
print(df.isnull().sum())

# Handle missing values
df = df.fillna({col: df[col].median() for col in num_cols})
df = df.fillna({col: df[col].mode()[0] for col in cat_cols})

# Missing values AFTER
print("\nMissing values after handling:\n")
print(df.isnull().sum())


```



*   Missing values were identified and handled systematically.
*   Numerical columns were imputed using median to reduce outlier impact.
*  Categorical columns were imputed using mode to preserve category         distribution.
*   After preprocessing, the dataset contains no missing values and is ready for EDA.

# **Correlation Matrix**

```python
important_cols = [
    'SalePrice',
    'OverallQual',
    'GrLivArea',
    'TotalBsmtSF',
    'GarageCars',
    'FullBath',
    'YearBuilt',
    'LotArea'
]

plt.figure(figsize=(10, 8))
sns.heatmap(
    df[important_cols].corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    square=True,
    cbar_kws={'shrink': 0.8}
)
plt.title("Correlation Matrix of Key Features", fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()


```



*   SalePrice shows strong positive correlation with OverallQual and GrLivArea.
*   GarageCars and TotalBsmtSF also have moderate positive relationships with price.
*   LotArea shows a weaker correlation, indicating land size alone is not a strong price driver.
*   Focusing on key variables provides clearer insights into factors influencing house prices.
